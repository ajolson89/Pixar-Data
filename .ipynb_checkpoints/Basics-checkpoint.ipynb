# Pixar Movies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

pixar_movies = pd.read_csv("PixarMovies.csv")
pixar_movies.shape
pixar_movies.head(len(pixar_movies))

#pixar_movies.dtypes
#pixar_movies.describe()
```

# Clean up Dataframe

```python
pixar_movies["Domestic %"] = pixar_movies["Domestic %"].str.rstrip("%").astype("float")
pixar_movies["International %"] = pixar_movies["International %"].str.rstrip("%").astype("float")

pixar_movies["IMDB Score"] = pixar_movies["IMDB Score"] * 10
filtered_pixar = pixar_movies.iloc[0:len(pixar_movies)-1]
filtered_pixar.set_index(["Movie"], inplace = True)
filtered_pixar
```

# Move rating by Critic Site

```python
critics_reviews = filtered_pixar[['RT Score', 'IMDB Score', 'Metacritic Score']]
critics_reviews.plot(figsize=(10,6))
plt.show()
```

In general it appears as though the Rotten Tomatoes site gives a higher rating than the other critic sites. 

# Boxplot showing critic data

```python
critics_reviews.plot(kind="box", figsize=(9,5))
plt.show()
```

This confirms our hypothesis from the line plot. Rotten Tomatoes has the highest average of the three review sites. 

# Bar plot showing Domestic/International Split

```python
revenue_proportions = filtered_pixar[["Domestic %", "International %"]]
revenue_proportions.plot(kind = "bar", stacked = True)
plt.legend(loc = 'upper right')
plt.show()
```

The movie list is also shown with oldest on the left and newest on the rightt. It is clear from this plot that newer movies are recieving increasing proportion of revenue from international markets. 

# Barplot comparing oscars nominated and oscar won

```python
oscars = filtered_pixar[['Oscars Nominated', 'Oscars Won']]
oscars.plot(kind = 'bar')
plt.show()
```

There doesn't necessarily appear to be a trend between oscars won and oscars nominated. However the dataset is too small to do meaningful statics on. A preferred option would be to look at all movies produced and nominated for an oscar. Intuitively there should be a relationship between oscars nominated and oscars won. 

# Revenue vs Movie

```python
noindex_pixar = pixar_movies.iloc[0:len(pixar_movies)-1]
sns.set(style="whitegrid", color_codes=True)
sns.factorplot(x = "Movie", y = "Adjusted Domestic Gross", data = noindex_pixar, kind = 'strip', jitter = True, aspect = 2)
plt.xticks(rotation = 90)
```

There almost appears to be a cyclical process to revenue and movie. Since the movies are listed in chronological order, perhaps this shows that there are movies which Pixar spends a large amount of time on, which are successes, and then produces a few lower quality movies which do not generate as much revenue. For instance, it took Pixar over a decade to produce Toy Story 3 after Toy Story 2 came out, this shows a large level of work went into the movie. It would be interesting to see if critic score relates to revenue. 

# Revenue vs Critic Score

```python
sns.jointplot(x = "Adjusted Domestic Gross", y="Metacritic Score", data = noindex_pixar)
```

In general there appears to be a linear trend associated with revenue and critic score, which makes sense. The higher the critic review the more the movie made in general.
