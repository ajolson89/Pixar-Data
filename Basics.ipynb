```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

pixar_movies = pd.read_csv("PixarMovies.csv")
pixar_movies.shape
pixar_movies.head(len(pixar_movies))

#pixar_movies.dtypes
#pixar_movies.describe()
```

```python
pixar_movies["Domestic %"] = pixar_movies["Domestic %"].str.rstrip("%").astype("float")
pixar_movies["International %"] = pixar_movies["International %"].str.rstrip("%").astype("float")

pixar_movies["IMDB Score"] = pixar_movies["IMDB Score"] * 10
filtered_pixar = pixar_movies.iloc[0:len(pixar_movies)-1]
filtered_pixar.set_index(["Movie"], inplace = True)
filtered_pixar
```

```python
critics_reviews = filtered_pixar[['RT Score', 'IMDB Score', 'Metacritic Score']]
critics_reviews.plot(figsize=(10,6))
plt.show()
```

```python
critics_reviews.plot(kind="box", figsize=(9,5))
plt.show()
```

```python
revenue_proportions = filtered_pixar[["Domestic %", "International %"]]
revenue_proportions.plot(kind = "bar", stacked = True)
plt.legend(loc = 'upper right')
plt.show()
```

```python
oscars = filtered_pixar[['Oscars Nominated', 'Oscars Won']]
oscars.plot(kind = 'bar')
plt.show()
```

```python
noindex_pixar = pixar_movies.iloc[0:len(pixar_movies)-1]
sns.set(style="whitegrid", color_codes=True)
sns.factorplot(x = "Movie", y = "Adjusted Domestic Gross", data = noindex_pixar, kind = 'strip', jitter = True, aspect = 2)
plt.xticks(rotation = 90)
```

```python
sns.jointplot(x = "Adjusted Domestic Gross", y="Metacritic Score", data = noindex_pixar)
```

```python
sns.boxplot(x = "Oscars Won", y = "Metacritic Score", data = noindex_pixar)
```
